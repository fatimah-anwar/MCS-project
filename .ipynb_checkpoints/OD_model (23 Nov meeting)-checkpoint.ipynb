```python
import mesa

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
from OD_model.model import opinionsModel
from OD_model.model import opinion_avg
```

## Peer trust only

```python
model = opinionsModel(N = 50, inst_info = -1)

while (True):
    prev_avg = opinion_avg(model)
    model.step()
    new_avg = opinion_avg(model)
    
    if abs(new_avg - prev_avg) < 0.00001:
        break
        
model_data = model.datacollector.get_model_vars_dataframe()
agents_data = model.datacollector.get_agent_vars_dataframe()
```

```python
sns.lineplot(x="Step", y="opinion", data=agents_data)

plt.xlabel("Time Step")
plt.ylabel("Opinion")
plt.show()
```

```python
for i in range(model.num_agents):
    sns.lineplot(x = "Step", y = "opinion", data = agents_data.query(f"AgentID == {i}"), color = 'grey' , alpha = 0.1)

plt.xlabel("Time Step")
plt.ylabel("Opinion")
plt.show()
```

```python
variance_data = agents_data.groupby("Step")["opinion"].var().reset_index()

sns.lineplot(x="Step", y="opinion", data=variance_data)

plt.xlabel("Time Step")
plt.ylabel("Variance")
plt.show()
```

```python
variance_data = agents_data.groupby("Step")["opinion"].var().reset_index()

sns.lineplot(x="Step", y="opinion", data=variance_data)

plt.xlabel("Time Step")
plt.ylabel("Variance")
plt.show()
```

Variance is a statistical measure that describes how much the individual opinions deviate from the average opinion at a given time step. Higher variance indicates greater spread or dispersion of opinions among the agents.

this plot helps visualize the dynamics of opinion variance among agents in the model. It provides insights into how opinions are spreading or converging over time, allowing to observe patterns or trends in the evolution of opinions within the simulated environment.

The alternating variance values between 0.11 and 0.13 suggest that, on average, the opinions of individual agents are not changing significantly over time. Here's an interpretation based on the variance values:

- Stability: The relatively consistent variance values (between 0.11 and 0.13) indicate that the spread or dispersion of opinions among the agents is not experiencing large fluctuations. In other words, the system appears to be stable with respect to opinion diversity.
- Consistency: If the variance values were consistently close to zero, it would suggest that the opinions of all agents are very close to each other, indicating a high level of consensus or uniformity. On the other hand, if the variance values were consistently high, it would indicate a high degree of disagreement or divergence among agents.
- Changes in Opinion: The alternating pattern might suggest that there are periods of time during which agents' opinions are relatively stable (resulting in lower variance) and other periods during which opinions might be more diverse or changing (resulting in higher variance).

**opinion diversity is stable + opinions of all agents are very close to each other + alternating pattern**
