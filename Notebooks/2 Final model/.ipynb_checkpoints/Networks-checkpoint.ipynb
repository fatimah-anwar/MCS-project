### 1. Regular lattice

```python
itera = 50
population = 100
```

```python
params = {
    "N": population,
#     "h": 5,
#     "w": 5,
#     "collect_model_data" : False,
#     "collect_agent_data" : False,
    "alpha" : 1,
    "network": "regular",
    "net_var2_m" : np.arange(10 , 100 , 10),
#     "net_var2_m" : 10
}

results = mesa.batch_run(
    myModel,
    parameters = params,
    
    iterations = itera,  
    max_steps = 100,  
    
    number_processes = 1,
    data_collection_period = 1,
    display_progress = True,
)

regular_network = pd.DataFrame(results)
```

```python
regular_network.query("RunId == 9 and Step == 1")['online_neighbors'].value_counts()
```

```python
dataset = regular_network

# plt.figure(figsize=(4,  3))

for pos in ['right', 'top', 'bottom', 'left']: 
    plt.gca().spines[pos].set_visible(False) 

sns.boxplot(x = 'net_var2_m' , y = 'mobility_rate' , data = dataset.query("Step == 100"),
           width = .3, linewidth = .8, fliersize = 0 , palette = "pastel") 

plt.ylim(0,1)
plt.xlabel("network")
plt.ylabel("Mobility rate")

plt.show()
```

```python
dataset = regular_network

# plt.figure(figsize=(4,  3))
for pos in ['right', 'top', 'bottom', 'left']: 
    plt.gca().spines[pos].set_visible(False) 

sns.lineplot(x = "Step" , y = "mobility_rate" , hue = "net_var2_m" ,data = dataset.query('Step < 31').groupby(["Step", "net_var2_m"])['mobility_rate'].mean().reset_index()) 

plt.ylim(0,1)
plt.xlabel("Time step")
plt.ylabel("Agents' mobility rate")
# plt.legend(bbox_to_anchor=(1, .86), ncol=1)

plt.show()
```
